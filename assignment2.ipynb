```python
sns.scatterplot(data=adult_df, x='age', y='capitalGain', hue='race', style='race', s=100)
plt.title('Scatterplot of Age vs. Capital Gain')
plt.xlabel('Age')
plt.ylabel('Capital Gain')
plt.legend(title='Race')
plt.show()
```
